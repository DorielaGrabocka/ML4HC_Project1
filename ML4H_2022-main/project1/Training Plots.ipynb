```python
import matplotlib.pyplot as plt
import json

from matplotlib.pyplot import figure
```

```python
path_to_train_hist = "CnnResidual_PTBDB/history.json"

with open(path_to_train_hist, 'r') as file:
    history = json.load(file)
    
```

```python
train_loss, val_loss = [], []
val_acc = []
val_f1 = []

for ep in history:
    train_loss.append(ep["train_loss"])
    val_loss.append(ep["valid_loss"])
    
    val_acc.append(ep["valid_acc"])
    val_f1.append(ep["skorch_f1_score"])
    
```

```python
plt.figure(figsize=(16,9))
plt.plot(train_loss)
plt.plot(val_loss)
plt.xlabel("Epoch")
plt.ylabel("Loss value")
plt.title("Residual CNN on PTBDB dataset")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.plot(val_acc)
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Residual CNN on PTBDB dataset")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.plot(val_f1)
plt.xlabel("Epoch")
plt.ylabel("Validation F1 score")
plt.title("Residual CNN on PTBDB dataset")
plt.show()
```
